```python
# Importing the pandas library for data manipulation
import pandas as pd

# Importing matplotlib and seaborn for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the iris dataset from seaborn library
iris = sns.load_dataset('iris')

# Creating a pandas DataFrame called 'dataset' from the iris dataset
dataset = pd.DataFrame(iris)
```

```python
# Displaying the first 10 rows of the dataset to examine the data structure
dataset.head(10)
```

```python
# Printing the mean values of all of the numeric features
print(dataset.mean(numeric_only=True).to_string(header=False))
```

```python
# Printing the median values of all of the numeric features
print(dataset.median(numeric_only=True).to_string(header=False))
```

```python
# Printing the mode values of all of the numeric features
print(dataset.mode(numeric_only=True))
```

```python
# Printing the variance values of all of the numeric features
print(dataset.var(numeric_only=True).to_string(header=False))
```

```python
# Printing the standard deviation values of all of the numeric features
print(dataset.std(numeric_only=True).to_string(header=False))
```

```python
# Creating the main plot
plt.figure(figsize=(10,10))

# Subplot histogram for Sepal Length
plt.subplot(2,2,1)
sns.histplot(dataset['sepal_length'], kde=True, color='blue')
plt.title('Sepal Length', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Sepal Width
plt.subplot(2,2,2)
sns.histplot(dataset['sepal_width'], kde=True, color='red')
plt.title('Sepal Width', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Petal Length
plt.subplot(2,2,3)
sns.histplot(dataset['petal_length'], kde=True, color='green')
plt.title('Petal Length', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Petal Width
plt.subplot(2,2,4)
sns.histplot(dataset['petal_width'], kde=True, color='orange')
plt.title('Petal Width', fontsize=16)
plt.ylabel('Frequency')

# Adjusting the plot gaps
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Showing the grids
plt.grid(True)

# Showing the main plot
plt.show()
```

```python
# Creating the main plot
plt.figure(figsize=(12, 12))

# Subplot boxplot for Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=dataset, color='blue')
plt.title('Sepal Length', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(True)

# Subplot boxplot for Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=dataset, color='red')
plt.title('Sepal Width', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Sepal Width')
plt.grid(True)

# Subplot boxplot for Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=dataset, color='green')
plt.title('Petal Length', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.grid(True)

# Subplot boxplot for Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=dataset, color='orange')
plt.title('Petal Width', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.grid(True)

# Adjusting the plot gaps
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Showing the main plot
plt.show()
```

```python
# Initiating the pair plot for complete scatter plot
# for all pari variables
sns.pairplot(dataset, hue='species', height=3)

# Showing the plot
plt.show()
```

```python
# Dropping the last column, species
features = dataset.drop(columns=['species'])

# Printing the numerical version of correlation matrix
print(features.corr())

# Presenting the CM to heat map
sns.heatmap(features.corr(), annot=True)

# Showing the heat map
plt.show()
```
